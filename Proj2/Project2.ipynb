## MAE

```python
torch.manual_seed(0)
lr = 1e-6
model = Sequential([Linear(2,25, lr),Tanh(),Linear(25,25, lr),Tanh(),Linear(25,25, lr),Tanh(),Linear(25, 2, lr), Sigmoid()])
train_set, train_labels = generate_dataset(1000)
test_set, test_labels = generate_dataset(1000)
losses, training_losses, test_losses = train(model, train_set, train_labels, 900, 25, MAE_loss, test_set, test_labels)
```

```python
print(get_errors(model, train_set, train_labels))
print(get_errors(model, test_set, test_labels))
print(get_metrics(model, train_set, train_labels))
avg_training_loss = sum(losses)/len(losses)
print(avg_training_loss)
```

```python
plot_loss_wrt_epochs(training_losses, test_losses)
```
